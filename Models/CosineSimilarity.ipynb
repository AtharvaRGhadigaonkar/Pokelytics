#### Data reading

```python
Kanto = pd.read_csv("../Data/Kanto.csv").drop("Unnamed: 0",axis=1)
Johto = pd.read_csv("../Data/Johto.csv").drop("Unnamed: 0",axis=1)
```

#### Rough Work

```python
movelist=[]
for x in Kanto["moves"]:
    words=x.split("'")
    for j in words:
        if "[" in j or "," in j or "]" in j:
            pass
        else:
            movelist.append(j)
```

```python
attacklist=[]
typelist=[]
for x in range(len(movelist)):
    if x %2==0:
        attacklist.append(movelist[x])
    else:
        typelist.append(movelist[x])
```

```python
attacklist =list(set(attacklist))
typelist =list(set(typelist))
```

```python
ayo =typelist
```

```python
Kanto.columns
```

```python
typelistad = typelist + ['hp', 'attack','defense', 'special-attack', 'special-defense', 'speed']
```

```python
typelist
```

```python
df = pd.DataFrame(columns=[attacklist])
df2 = pd.DataFrame(columns=[typelistad])
```

```python
for y in range(len(Kanto["name"])):
    arr =[]
    # arr.append(Kanto["name"][y])
    for x in range(len(attacklist)):
        if(attacklist[x] in (Kanto["moves"][y])):
            arr.append(1)
        else:
            arr.append(0)
    df.loc[y]= arr
```

```python
for y in range(len(Kanto["name"])):
    arr =[]
    # arr.append(Kanto["name"][y])
    for x in range(len(ayo)):
        if(typelist[x] in (Kanto["type1"][y],Kanto["type2"][y])):
            arr.append(1)
        else:
            arr.append(0)
    for z in ['hp', 'attack','defense', 'special-attack', 'special-defense', 'speed']:
            arr.append(Kanto[z][y])
    # print(arr)
    df2.loc[y]= arr
```

#### Cosine

```python
df1=df
```

```python
scores= cosine_similarity(df1)
```

```python
scores.shape
```

```python
sorted(list(enumerate(scores.tolist())),key=lambda x:x[1],reverse=True)
```

```python
def recommend_moves(name):
        index = Kanto[Kanto["name"]==name].index[0]
        distance = scores[index]
        items = sorted( list(enumerate(distance)) ,key=lambda x:x[1],reverse=True)[1:6]
        for x in items:
            print(Kanto["name"][x[0]])
            # print(x[0])
```
